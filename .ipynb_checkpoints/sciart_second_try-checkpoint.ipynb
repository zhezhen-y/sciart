```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
```

```python
diag = np. eye(30,30)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.plot(diag, linestyle='--', marker='o', linewidth=4, markersize=12)
ax.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.set_prop_cycle(color=['red','orange','yellow','green','blue','purple'])
ax.plot(diag, marker='o', linewidth=2, markersize=12)
ax.axis('off')
plt.show()
```

```python
diag = np. eye(50,50)
fig, ax = plt.subplots(figsize=(20,15))
ax.set_prop_cycle(color=['red','orange','yellow','green','blue','purple'])
ax.plot(diag, marker='o', linewidth=2, markersize=12)
ax.axis('off')
plt.show()
```
